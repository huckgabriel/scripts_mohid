```python
import pandas as pd
```

```python
# rutas / path
path_anidado = ''
path_padre = ''
```

```python
def clean_file_srh(file_name):
    with open(file_name, 'r') as f:
        lines = f.readlines()
    print(lines[0])

    if 'Time' in lines[0]:
        # 1- Remove the first 10 lines
        lines = lines[10:] # Corrected variable assignment

        # 2- Remove line with "<Begintimeserie>"
        lines = [line for line in lines if "<BeginTimeSerie>" not in line]

        # 3- Delete block between <endtimeserie> and <endresidual> (both included)
        result = []
        for line in lines: # Iterating over lines instead of line
            if "<EndTimeSerie>" in line: # Corrected tag name
                continue
            if "<BeginResidual>" in line:
                continue
            if "<EndResidual>" in line:
                continue
            if line.strip() == "":
                continue
            result.append(line)

        # rewrite file called f
        with open(file_name, 'w') as f:
            f.writelines(result) # result is already a list of strings

    else:
        print('The file is already clean')
```

```python
file_name = path_padre + '/' + 'E1 - copia.srh'
clean_file_srh(file_name)
```
